```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = {
    "Age":[19,21,20,23,31,22,35,23,64,30,67,35,58,24,37,22,35,20,52,35],
    
    "Annual Income":[15,15,16,16,17,17,18,18,19,19,20,20,21,21,22,22,23,23,24,24],
    
    "Spending Score":[39,81,6,77,40,76,6,94,3,72,14,99,15,77,13,79,35,66,29,98]
}

df = pd.DataFrame(data)
df
```

```python
X = df[['Age','Annual Income','Spending Score']]
```

```python
kmeans = KMeans(n_clusters=5, random_state=0)
```

```python
df['Cluster'] = kmeans.fit_predict(X)
df
```

```python
centroids = kmeans.cluster_centers_

print("Centroids:")
print(centroids)
```

```python
plt.scatter(df['Annual Income'], df['Spending Score'], 
            c=df['Cluster'], cmap='rainbow')

plt.scatter(centroids[:,1], centroids[:,2], 
            s=200, c='black', marker='X')

plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Mall Customer Segmentation")

plt.show()
```
