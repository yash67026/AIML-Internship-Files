```python
import nltk
import random
from nltk.tokenize import word_tokenize
from collections import defaultdict, Counter
nltk.download('punkt')
```

```python
corpus = """
Artificial intelligence is transforming the world.
Machine learning models learn from data and improve performance.
Natural language processing helps computers understand human language.
Deep learning is a subset of machine learning.
AI is widely used in automation and cloud computing.
"""
```

```python
tokens = word_tokenize(corpus.lower())
```

```python
model = defaultdict(list)
for i in range(len(tokens) - 1):
    model[tokens[i]].append(tokens[i + 1])
```

```python
def predict_next(word):
    next_words = model[word]
    
    if not next_words:
        return random.choice(tokens)
    
    freq = Counter(next_words)
    total = sum(freq.values())
    probs = {w: freq[w] / total for w in freq}
    
    return random.choices(list(probs.keys()), weights=probs.values())[0]
```

```python
def generate_sentence(start_word, length=10):
    sentence = [start_word.lower()]
    
    for _ in range(length - 1):
        next_word = predict_next(sentence[-1])
        sentence.append(next_word)
    
    return " ".join(sentence)
```

```python
user_input = input("Enter a starting word: ")
output = generate_sentence(user_input, 10)

print("\nGenerated Sentence:")
print(output)
```
