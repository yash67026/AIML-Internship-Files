```python
# INMPORT LIBRARIES
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#create dataset
data = {
    "Age" : [22,23,24,26,31, 32, 47, 45,49],
    "Income":[25,28,26,90,88,95,60,65,62]
}
df = pd.DataFrame(data)
print (df)
```

```python
#Visualize data
import matplotlib.pyplot as plt

plt.scatter(df["Age"], df["Income"])
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
```

```python
#apply k-means 
kmeans = KMeans(n_clusters =3)
kmeans.fit(df)
```

```python
#get cluster labels
labels = kmeans.labels_
print(labels)
```

```python
'''
cluster 1 = young
cluster 2 = high income
cluster 3 = 
'''
```

```python
#show centroids
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
plt.scatter(df["Age"], df["Income"], c = labels)
plt.scatter(centroids[:,0], centroids[:,1], marker = 'X', s = 200)
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("K-Means Clustering of Age vs Income")
plt.show()
```
