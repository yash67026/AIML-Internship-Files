```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
data = {
    'Weather': ['Sunny','Sunny','Overcast','Rain','Rain','Rain','Overcast','Sunny','Sunny','Rain'],
    'Temperature': ['Hot','Hot','Hot','Mild','Cool','Cool','Mild','Cool','Mild','Mild'],
    'Humidity': ['High','High','High','High','Normal','Normal','Normal','High','Normal','Normal'],
    'Wind': ['Weak','Strong','Weak','Weak','Weak','Strong','Strong','Weak','Weak','Strong'],
    'PlayOutside': ['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes']
}

df = pd.DataFrame(data)
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in df.columns:
    df[column] = le.fit_transform(df[column])
df
```

```python
X = df.drop('PlayOutside', axis=1)
y = df['PlayOutside']
```

```python
model = DecisionTreeClassifier()
model.fit(X, y)
```

```python
plt.figure(figsize=(10,6))
tree.plot_tree(
    model,
    feature_names=['Weather','Temperature','Humidity','Wind'],
    class_names=['No','Yes'],
    filled=True
)
plt.show()
```
