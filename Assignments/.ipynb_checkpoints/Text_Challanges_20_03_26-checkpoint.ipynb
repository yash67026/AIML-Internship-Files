```python
data = [
    "omg this movie is lit 🔥🔥",
    "I cant beleive this happend",
    "LOL that was sooo funny 😂",
    "brb going 2 market",
    "gud mrng frnds 😊",
    "This is awsm!!!",
    "idk what u mean",
    "heyyy!!! whats up???",
    "I luv this song ❤️",
    "thnx 4 ur help",
    "pls send d file asap",
    "sooo tired rn 😴",
    "w8 for me",
    "y r u late??",
    "dis is not gud",
    "haha thats gr8 😄",
    "c u soon",
    "omg!!! whattt is this??",
    "im sooo happpy!!!",
    "okkkk done 👍"
]

df = pd.DataFrame(data, columns=["Sentence"])
df
```

```python
def to_lower(text):
    return text.lower()
```

```python
def remove_punctuation(text):
    punctuations = "!?,.:-;()[]{}'\""
    for p in punctuations:
        text = text.replace(p, "")
    return text
```

```python
slang_dict = {
    "omg": "oh my god",
    "lol": "laughing out loud",
    "brb": "be right back",
    "idk": "i dont know",
    "u": "you",
    "ur": "your",
    "thnx": "thanks",
    "pls": "please",
    "rn": "right now",
    "gr8": "great",
    "w8": "wait",
    "c": "see",
    "luv": "love",
    "gud": "good",
    "mrng": "morning",
    "frnds": "friends",
    "dis": "this",
    "y": "why",
    "u" : "you"
}
```

```python
def replace_slang(text):
    words = text.split()
    new_words = []
    
    for word in words:
        if word in slang_dict:
            new_words.append(slang_dict[word])
        else:
            new_words.append(word)
            
    return " ".join(new_words)
```

```python
df["Processed"] = df["Sentence"].apply(to_lower)
df["Processed"] = df["Processed"].apply(remove_punctuation)
df["Processed"] = df["Processed"].apply(replace_slang)

df
```
