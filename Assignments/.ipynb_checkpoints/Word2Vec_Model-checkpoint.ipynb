```python
from gensim.models import Word2Vec
```

```python
sentences = [
    "college life is fun and exciting",
    "students attend classes every day",
    "friends make college life memorable",
    "we study machine learning and ai",
    "assignments must be completed on time",
    "teachers guide students in learning",
    "library is useful for studying",
    "group study helps in understanding",
    "exams are stressful but important",
    "projects improve practical knowledge",
    "campus is large and beautiful",
    "students participate in sports activities",
    "internships help in career growth",
    "learning new skills is important",
    "college events are enjoyable"
]
```

```python
tokenized_sentences = []
for sentence in sentences:
    tokenized_sentences.append(sentence.lower().split())
tokenized_sentences
```

```python
model = Word2Vec(
    sentences=tokenized_sentences,
    vector_size=50,
    window=3,
    min_count=1,
    workers=1
)
```

```python
vector = model.wv['college']
print("Vector for 'college':\n", vector)
```

```python
similar_words = model.wv.most_similar('college', topn=5)
print("Top 5 similar words to 'college':")
for word, score in similar_words:
    print(word, ":", score)
```
