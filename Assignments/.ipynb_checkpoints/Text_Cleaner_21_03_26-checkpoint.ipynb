# Build a Text Cleaner

```python
import pandas as pd
```

```python
data = [
    "This is a simple sentence!",
    "I am learning NLP, and it is fun.",
    "The movie was very good and interesting.",
    "This is not bad at all.",
    "He is going to the market."
]

df = pd.DataFrame(data, columns=["Sentence"])
df
```

```python
def to_lower(text):
    return text.lower()
```

```python
def remove_punctuation(text):
    punctuations = "!?,.:-;()[]{}'\""
    for p in punctuations:
        text = text.replace(p, "")
    return text
```

```python
stopwords = [
    "is", "am", "are", "the", "and", "to", "a", "of", "in", "it", "was"
]
```

```python
def remove_stopwords(text):
    words = text.split()
    filtered_words = []
    
    for word in words:
        if word not in stopwords:
            filtered_words.append(word)
    
    return " ".join(filtered_words)
```

```python
df["Cleaned"] = df["Sentence"].apply(to_lower)
df["Cleaned"] = df["Cleaned"].apply(remove_punctuation)
df["Cleaned"] = df["Cleaned"].apply(remove_stopwords)
df
```

```python
for i in range(len(df)):
    print("Original :", df["Sentence"][i])
    print("Cleaned  :", df["Cleaned"][i])
    print("-"*40)
```
